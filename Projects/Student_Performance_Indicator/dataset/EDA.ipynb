### Student Performance Indicator

#### Problem Statement

* This project understands how the sutdents performance (test scores) is affected by other variables such as Gender, Ethnicity, Parental level of education, Lunch and Test preparation course.

```python
#exploring each features

print(f"Categories in gender: {df['gender'].unique()}\n\n")

print(f"Categories in race_ethnicity: {df['race_ethnicity'].unique()}\n\n")

print(f"Categories in parental_level_of_education: {df['parental_level_of_education'].unique()}\n\n")

print(f"Categories in lunch: {df['lunch'].unique()}\n\n")

print(f"Categories in test_preparation_course: {df['test_preparation_course'].unique()}\n\n")
```

```python
#statistics check

df.describe(include="all").T
```

```python
#number of categorical and numerical features

cat_features = [col for col in df.columns if df[col].dtype == "str"]
print(f"Catgorical Features: {cat_features}")

num_features = [col for col in df.columns if df[col].dtype != "str"]
print(f"Numerical Features: {num_features}")
```

```python
#feature extraction- adding total_score and avg_score features

df['total_score'] = df['math_score']+df['reading_score']+df['writing_score']
df['avg_score'] = df['total_score']/3
```

```python
df.head(2)
```

```python
math_full = df[df['math_score']==100]['avg_score'].count()
reading_full = df[df['reading_score']==100]['avg_score'].count()
writing_full = df[df['writing_score']==100]['avg_score'].count()

print(f"Number of students with full score in maths: {math_full}")
print(f"Number of students with full score in reading: {reading_full}")
print(f"Number of students with full score in writing: {writing_full}")
```

```python
math_lessthan_20 = df[df['math_score']<=20]['avg_score'].count()
reading_lessthan_20 = df[df['reading_score']<=20]['avg_score'].count()
writing_lessthan_20 = df[df['writing_score']<=20]['avg_score'].count()

print(f"Number of students with less than 20 score in maths: {math_lessthan_20}")
print(f"Number of students with less than 20 in reading: {reading_lessthan_20}")
print(f"Number of students with less than 20 in writing: {writing_lessthan_20}")
```

```python
#visualization

plt.subplots(1,2, figsize=(10,6))

plt.subplot(1,2,1)
sns.histplot(x='avg_score',data=df, kde=True, color='y')


plt.subplot(1,2,2)
sns.histplot(x='avg_score',data=df, kde=True, hue='gender', palette='viridis')
plt.show()
```
