### Decision Tree Classifier

```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
df.head()
```

```python
y = iris['target']
y
```

```python
# training and test datasets

x_train, x_test, y_train, y_test = train_test_split(df.values, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#model training

decision_tree = DecisionTreeClassifier()
decision_tree.fit(x_train, y_train)
```

```python
#model prediction

y_pred = decision_tree.predict(x_test)
y_pred
```

```python
#performance metrics 

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#visualization of decision tree

from sklearn.tree import plot_tree
plt.figure(figsize=(10,10))
plot_tree(decision_tree, filled=True)
plt.show()
```

* To avoid over-fitting we can use post-pruning and pre-pruning.
* post-pruning is used if dataset is small and in pre-pruning we tune hyperparameters.


```python
#using post-pruning here: we see model is able to differenciate at max_depth=2

pruned_classifier = DecisionTreeClassifier(max_depth=2)
pruned_classifier
```

```python
#training and prediction on pruned model

pruned_classifier.fit(x_train, y_train)
y_pred = pruned_classifier.predict(x_test)
```

```python
#performance metrics 

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#visualization of decision tree

from sklearn.tree import plot_tree
plt.figure(figsize=(6,6))
plot_tree(pruned_classifier, filled=True)
plt.show()
```
