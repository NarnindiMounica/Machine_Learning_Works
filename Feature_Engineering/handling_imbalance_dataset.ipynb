### Handling Imbalanced Dataset

* Upsampling
* Downsampling

```python
import numpy as np
import pandas as pd
```

```python
#creating imbalanced dataset

class_0_ratio = 0.9
n_samples = 1000

n_class_0_samples = int(1000 * class_0_ratio)
n_class_1_samples = n_samples - n_class_0_samples
```

```python
class_0_df = pd.DataFrame(
    {"feature_1": np.random.randn(n_class_0_samples),
    "feature_2": np.random.randn(n_class_0_samples),
    "target": [0]*n_class_0_samples }
)

class_1_df = pd.DataFrame(
    {"feature_1": np.random.randn(n_class_1_samples),
    "feature_2": np.random.randn(n_class_1_samples),
    "target": [1]*n_class_1_samples }
)

full_df = pd.concat([class_0_df, class_1_df], axis=0)
full_df.head()
```
