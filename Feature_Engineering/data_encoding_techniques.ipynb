### Data Encoding

#### 1.Nominal/One-Hot Encoding:

One hot encoding also known as nominal encoding is a technique used to represent categorical data as numerical data which is needed for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
#sample dataframe for demonstration

color_df = pd.DataFrame(
    {"color": ['red', 'blue', 'yellow', 'blue', 'blue', 'red', 'yellow']}
)
color_df
```

```python
ohe_obj = OneHotEncoder()
color_data=ohe_obj.fit_transform(color_df[['color']]).toarray()
```

```python
cols=ohe_obj.get_feature_names_out()
```

```python
new_df = pd.DataFrame(data=color_data, columns=cols)
new_df
```

```python
#to encode new data

ohe_obj.transform([['blue']]).toarray()
```

#### Using tips dataset

```python
import seaborn as sns
tips_df = sns.load_dataset('tips')
```

#### 2.Label Encoding:

Label encoding involves assigning a unique numerical label to each category in the variable. The labels are usually assigned in alphabetical order or based on the frequency of the categories.

```python
color_df
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

label_encoder.fit_transform(color_df)
```

```python
#applying encoding on new data

label_encoder.transform(['blue'])
```

#### 3.Ordinal Encoding:

It is used to encode categorical features that has intrinsic order or ranking. In this technique, each category is assigned a numerical value based on it's position in the order.

```python
color_df
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder(categories=[['yellow', 'blue', 'red']])
ordinal_encoder.fit_transform(color_df[['color']])
```

```python
#to encode new data

ordinal_encoder.transform([['blue']])
```

### 4.Target Guided Ordinal Encoder

* It is a technique used to encode categorical variables based on their relationship with the target variables. This encoding technique is useful when we have a categorical variable with large number of unique categories, and we want to use this variable as a feature in our model.

* In this technique we replace each category in the categorical variable with a numerical value based on the mean or median of the target variable for that category.


```python
df = pd.DataFrame(
    {"country": ["China", "India", "Japan", "Japan", "India", "China", "India"], 
     "target":[6, 9, 9, 7, 9, 9, 9 ]}
)

df
```

```python
country_dict=df.groupby('country')['target'].mean().to_dict()
country_dict
```

```python
df['country_encoded'] = df['country'].map(country_dict)
df['country_encoded']
```
