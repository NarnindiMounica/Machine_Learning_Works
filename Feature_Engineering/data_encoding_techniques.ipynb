### Data Encoding

#### 1.Nominal/One-Hot Encoding:

One hot encoding also known as nominal encoding is a technique used to represent categorical data as numerical data which is needed for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
#sample dataframe for demonstration

color_df = pd.DataFrame(
    {"color": ['red', 'blue', 'yellow', 'blue', 'blue', 'red', 'yellow']}
)
color_df
```

```python
ohe_obj = OneHotEncoder()
color_data=ohe_obj.fit_transform(color_df[['color']]).toarray()
```

```python
cols=ohe_obj.get_feature_names_out()
```

```python
new_df = pd.DataFrame(data=color_data, columns=cols)
new_df
```

```python
#to encode new data

ohe_obj.transform([['blue']]).toarray()
```
