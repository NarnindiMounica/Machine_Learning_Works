### Handling Imbalanced Dataset Using SMOTE (Synthetic Minority Oversampling Technique)

* SMOTE is a technique used in machine learning to address imbalance dataset where the minority class has significantly fewer instances than the majority class.

* SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.


```python
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification
import seaborn as sns
import pandas as pd
```

```python
#to create imbalanced dataset

x, y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_clusters_per_class=1, weights=[0.9], random_state=42)
```

```python
df=pd.DataFrame(
    {"feature_1": x[:,0],
     "feature_2": x[:,1],
     "target":y}
)
```

```python
df.target.value_counts()
```

```python
sns.scatterplot(x=df['feature_1'],y=df['feature_2'], hue=df['target'])
```

```python
smote_obj = SMOTE()
x, y = smote_obj.fit_resample(X=df[['feature_1', 'feature_2']], y=df['target'])
```

```python
new_df = pd.concat([x,y], axis=1)
new_df.target.value_counts()
```

```python
sns.scatterplot(x=new_df['feature_1'], y=new_df['feature_2'], hue=new_df['target'])
```
