### 5 Number Summary & BoxPlot

```python
import numpy as np
import seaborn as sns
```

```python
#sample values

sample_list = np.random.choice(np.arange(10,100), 20)
sample_list = sample_list.tolist()
```

```python
min, q1, median, q3, max = np.quantile(sample_list, [0, 0.25, 0.5, 0.75, 1])
print(f"Minimum value: {min}\nQ1 value: {q1}\nMedian value: {median}\nQ3 value: {q3}\nMaximum value: {max}")
```

```python
sns.boxplot(sample_list)
```

```python
#adding some outliers in data for visualization

sample_list.extend([300, 289, -20, -50])
sample_list
```

```python
sns.boxplot(sample_list)
```
